```python
import yfinance as yf
import pandas as pd
import random
import datetime
```

```python
ticker_list = ["AAPL", "MSFT", "LLY", "XOM", "UNH", "WMT", "JPM", "JNJ", "PG", "HD", "CVX", "ORCL", "MRK", "KO", "ADBE", "COST", "PEP", "BAC", "MCD", "TMO", "CMCSA", "PFE", "DHR", "ABT", "AMD", "INTC", "WFC", "DIS", "TXN", "COP", "NKE", "CAT", "VZ", "AMGN", "NEE", "IBM", "LOW", "UNP", "HON", "GE", "SPGI", "BA", "BMY", "AMAT", "DE", "AXP", "SYK", "MDT", "RTX", "LMT", "SCHW", "T", "TJX", "ADP", "MMC", "ADI", "CVS", "ETN", "LRCX", "SLB", "CI", "C", "PGR", "MU", "EOG", "MO", "SO", "BDX", "FI", "DUK", "ITW", "AON", "SHW", "APD", "CDNS", "NOC", "KLAC", "WM", "FDX", "CSX", "CL", "GD", "MNST", "OXY", "HUM", "EMR", "MMM", "TGT", "USB", "CTAS", "VLO", "HES", "PH", "AJG", "ECL", "PNC", "F", "MSI", "PSA", "NSC"]
```

```python
df_first =yf.download("AAPL", start="1990-01-01", end="2019-12-31")
data_df = pd.DataFrame(index=df_first.index, data = {'date':df_first.index})
data_df['date']=data_df['date'].apply(lambda a:pd.to_datetime(a).date())
for ticker in ticker_list:
    df =yf.download(ticker, start="1990-01-01", end="2019-12-31")
    data = pd.DataFrame(index=df_first.index, data = {ticker:df.Close})
    data_df = data_df.merge(data, left_index=True, right_index=True, how='outer')
```

```python
data_df.to_csv('result'+'.csv', index=False)
```
